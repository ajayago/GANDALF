```python
confident_pseudolabels_df = pseudolabels_df[pseudolabels_df.pseudolabels_binary != -1]
confident_pseudolabels_df.reset_index(inplace=True)
confident_pseudolabels_df_idx = confident_pseudolabels_df.index 
```

```python
confident_pseudolabels_df
```

```python
cl_auc_df = pd.read_csv("/data/ajayago/papers_data/systematic_assessment/raw/response_files/ccle_auc.csv")
cl_auc_df
```

```python
auc_values = confident_pseudolabels_df.merge(cl_auc_df, on=["sample_id", "drug_name"], how="left")
auc_values
```

```python
auc_reduced = auc_values[~auc_values["auc"].isna()]
auc_reduced
```

```python
auc_reduced["pseudolabels_binary"].value_counts()
```

```python
spearmanr(auc_reduced["pseudolabels"], 1-auc_reduced["auc"])
```
