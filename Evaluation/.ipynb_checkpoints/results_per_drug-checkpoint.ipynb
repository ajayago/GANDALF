```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score
```

```python
drugs = ["CISPLATIN", "FLUOROURACIL", "TEMOZOLOMIDE", "PACLITAXEL", "GEMCITABINE"]
```

```python
results_folder = "/data/ajayago/papers_data/DiffDRP_v7/run_files/saved_model_annotated_mutations/working_pseudolabeling_per_drug/"
```

```python
fold = 0
res_df_fold0 = pd.DataFrame()
combined_df_fold0 = pd.DataFrame()
for d in drugs:
    # print(d)
    res_df = pd.read_csv(f"{results_folder}prediction_patients_val_corr_2A_ALL_fold{fold}_tuned4{d}.csv", index_col = 0)
    # print(res_df.shape)
    auroc = roc_auc_score(res_df["y_true"], res_df["y_pred"])
    auprc = average_precision_score(res_df["y_true"], res_df["y_pred"])
    # print(f"AUROC = {auroc}, AUPRC = {auprc}")
    combined_df_fold0 = pd.concat([combined_df_fold0, res_df], ignore_index=True)
    res_df_fold0 = res_df_fold0._append({"drug": d, "auroc": auroc, "auprc": auprc}, ignore_index=True)

print("For combined df")
print(combined_df_fold0.shape)
auroc = roc_auc_score(combined_df_fold0["y_true"], combined_df_fold0["y_pred"])
auprc = average_precision_score(combined_df_fold0["y_true"], combined_df_fold0["y_pred"])
print(f"AUROC = {auroc}, AUPRC = {auprc}")
print(res_df_fold0)
```
