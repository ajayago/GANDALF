```python
import itertools
```

```python
from scipy.stats import pearsonr, spearmanr, kstest
```

```python
pseudolabels_df = pd.read_csv("/data/ajayago/papers_data/DiffDRP_v7/run_files/saved_model_annotated_mutations/pseudolabels_df_val_corr_2A_ALL_fold2.csv", index_col = 0)
```

```python
pseudolabels_df[pseudolabels_df.majority_vote_pseudolabels != -1]
```

```python
cl_aug_df = pd.read_csv("/data/ajayago/papers_data/DiffDRP_v7/run_files/saved_model_annotated_mutations/augmented_cl_clconditioned_uda_v2_vaeinput_val_corr_2A_ALL_fold2.csv", index_col = 0)
cl_aug_df
```

```python
cl_train_val = list(cl_aug_df.index)
```

```python
drug_fp = pd.read_csv("/data/ajayago/papers_data/systematic_assessment/raw/metadata/drug_morgan_fingerprints.csv", index_col = 0)
drugs_with_fp = list(drug_fp.index)
```

```python
possible_combos = list(itertools.product(cl_train_val, drugs_with_fp))
```

```python
possible_cl_drug_combinations_df = pd.DataFrame(possible_combos, columns = ["sample_id", "drug_name"])
possible_cl_drug_combinations_df
```

```python
confident_pseudolabels_df = pseudolabels_df[pseudolabels_df.majority_vote_pseudolabels != -1]
confident_pseudolabels_df_idx = confident_pseudolabels_df.index 
```

```python
subset = possible_cl_drug_combinations_df[possible_cl_drug_combinations_df.index.isin(confident_pseudolabels_df_idx)]
subset
```

```python
subset["pseudolabels"] = list(confident_pseudolabels_df["majority_vote_pseudolabels"])
subset
```

```python
cl_auc_df = pd.read_csv("/data/ajayago/papers_data/systematic_assessment/raw/response_files/ccle_auc.csv")
cl_auc_df
```

```python
auc_values = subset.merge(cl_auc_df, on=["sample_id", "drug_name"], how="left")
auc_values
```

```python
auc_reduced = auc_values[~auc_values["auc"].isna()]
auc_reduced
```

```python
auc_reduced["pseudolabels"].value_counts()
```

```python
spearmanr(auc_reduced["pseudolabels"], 1-auc_reduced["auc"]) # only fold 2 has samples with both labels, atleast enough to check correlation
```
