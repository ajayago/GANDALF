```python
fold = 0
```

```python
augmented_cl_df = pd.read_csv(f"/data/ajayago/papers_data/DiffDRP_v7/run_files/saved_model_annotated_mutations/augmented_cl_clconditioned_uda_v2_vaeinput_val_corr_2A_ALL_fold{fold}.csv", index_col = 0)
augmented_cl_df
```

```python
drug_fp = pd.read_csv("/data/ajayago/papers_data/systematic_assessment/raw/metadata/drug_morgan_fingerprints.csv", index_col = 0)
drug_fp
```

```python
metadata_df = pd.read_csv("/data/ajayago/papers_data/systematic_assessment/raw/metadata/labelled_ccle_metadata.csv", index_col = 0)
metadata_df
```

```python
metadata_df["OncotreeLineage"].value_counts()
```

```python
pseudolabels_df = pd.read_csv(f"/data/ajayago/papers_data/DiffDRP_v7/run_files/saved_model_annotated_mutations/working_fold{fold}/mtl_pseudolabel_v2/pseudolabel_df_fold{fold}.csv")
pseudolabels_df
```

```python
# load the fold based on sample_id - Note: cell lines have only 1 fold (fold 0)
with open("/data/ajayago/papers_data/systematic_assessment/input_types/annotated_mutations/Experiment2/SettingA//cell_lines_fold0_processed.pkl", "rb") as f:
    source_data = pickle.load(f)
```

```python
source_data["train"]
```

```python
from sklearn.manifold import TSNE
```

```python
suffixes = ["_piu_max", "_piu_sum", "_piu_mean", "_piu_count",
            "_lu_max", "_lu_sum", "_lu_mean", "_lu_count",
            "_ncu_max", "_ncu_sum", "_ncu_mean", '_ncu_count',
            "_pathogenic_max", "_pathogenic_sum", "_pathogenic_mean", "_pathogenic_count",
            "_vus_max", "_vus_sum", "_vus_mean", "_vus_count",
            "_benign_max", "_benign_sum", "_benign_mean", "_benign_count"
           ]
genes_7776 = []
for s in suffixes:
    for g in list(pd.read_csv(f"/data/ajayago/papers_data/systematic_assessment//raw/metadata/gene2ind.txt", header=None)[0]):
        genes_7776.append(f"{g}{s}")
```

```python
source_data["train"][source_data["train"].sample_id.isin(augmented_cl_df.index)].drop_duplicates(subset=["sample_id"])
```

```python
orig_cl_data = source_data["train"][source_data["train"].sample_id.isin(augmented_cl_df.index)].drop_duplicates(subset=["sample_id"])[genes_7776 + ["sample_id"]]
orig_cl_data
```

```python
orig_cl_data = orig_cl_data.merge(metadata_df[["ProfileID", "OncotreeLineage"]], left_on=["sample_id"], right_on=["ProfileID"])
orig_cl_data
```

```python
augmented_cl_df = augmented_cl_df.reset_index().merge(metadata_df[["ProfileID", "OncotreeLineage"]], left_on=["index"], right_on=["ProfileID"])
augmented_cl_df
```

```python
tsne_orig = TSNE(n_components=2, perplexity=10, random_state=21)
tsne_aug = TSNE(n_components=2, perplexity=10, random_state=21)
```

```python
tsne_np = tsne_orig.fit_transform(orig_cl_data[genes_7776])
```

```python
tsne_orig_df = pd.DataFrame(tsne_np, columns = ["tsne_comp1", "tsne_comp2"])
tsne_orig_df["ProfileID"] = orig_cl_data["ProfileID"]
tsne_orig_df["OncotreeLineage"] = orig_cl_data["OncotreeLineage"]
tsne_orig_df
```

```python
tsne_np = tsne_aug.fit_transform(augmented_cl_df.filter(regex="vae_*"))
```

```python
tsne_aug_df = pd.DataFrame(tsne_np, columns = ["tsne_comp1", "tsne_comp2"])
tsne_aug_df["ProfileID"] = augmented_cl_df["ProfileID"]
tsne_aug_df["OncotreeLineage"] = augmented_cl_df["OncotreeLineage"]
tsne_aug_df
```

```python
tsne_orig_df.OncotreeLineage.value_counts()
```

```python
lineages = ["Lung", "Lymphoid", "CNS/Brain", "Skin", "Esophagus/Stomach", "Bowel", "Breast"]
# lineages = ["Lymphoid", "Liver"]
```

```python
sns.scatterplot(data=tsne_orig_df[tsne_orig_df.OncotreeLineage.isin(lineages)], x="tsne_comp1", y="tsne_comp2", hue="OncotreeLineage")
```

```python
sns.scatterplot(data=tsne_aug_df[tsne_aug_df.OncotreeLineage.isin(lineages)], x="tsne_comp1", y="tsne_comp2", hue="OncotreeLineage")
```

```python
import umap
```

```python
import umap.plot
```

```python
mapper = umap.UMAP(random_state=42, n_components=2, n_neighbors=20, min_dist=1).fit(augmented_cl_df.filter(regex="vae_*"))
```

```python
umap.plot.points(mapper, labels=augmented_cl_df["OncotreeLineage"])
```

```python
fitted = umap.UMAP(random_state=42, n_components=2, n_neighbors=5, min_dist=0.001).fit_transform(augmented_cl_df.filter(regex="vae_*"))
fitted
```

```python
umap_aug_df = pd.DataFrame(fitted[:, :2], columns = ["umap_comp1", "umap_comp2"])
umap_aug_df["ProfileID"] = augmented_cl_df["ProfileID"]
umap_aug_df["OncotreeLineage"] = augmented_cl_df["OncotreeLineage"]
umap_aug_df
```

```python
sns.scatterplot(data=umap_aug_df[umap_aug_df.OncotreeLineage.isin(lineages)], x="umap_comp1", y="umap_comp2", hue="OncotreeLineage")
```

```python
fitted_orig = umap.UMAP(random_state=42, n_components=2, n_neighbors=5, min_dist=0.001).fit_transform(orig_cl_data[genes_7776])
fitted_orig
```

```python
umap_orig_df = pd.DataFrame(fitted_orig[:, :2], columns = ["umap_comp1", "umap_comp2"])
umap_orig_df["ProfileID"] = orig_cl_data["ProfileID"]
umap_orig_df["OncotreeLineage"] = orig_cl_data["OncotreeLineage"]
umap_orig_df
```

```python
sns.scatterplot(data=umap_orig_df[umap_orig_df.OncotreeLineage.isin(lineages)], x="umap_comp1", y="umap_comp2", hue="OncotreeLineage")
```
